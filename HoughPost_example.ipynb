- load `hough` package

```r
install_github("ChenPanxi/Statistical_Computing/HoughPost")
```

```r
library(HoughPost)
```

- lines

```r
coords_line = simulator(nline=5, npoint=20, nnoise=100, coordinate='cartesian', line_noise=TRUE, noise=TRUE)
```

```r
hs_line = hough_space(coords_line, thres=12, coordinate='cartesian')
```

```r
ht_line = hough_trans(coords_line, hs_line, coordinate='cartesian')
```

- circle

```r
circle_noise = simulator(nline=3, npoint=150, coordinate='circle', line_noise=TRUE, noise=TRUE)
```

```r
hs_circle = hough_space(circle_noise, steps=90, thres=200, coordinate='circle')
```

```r
ht_circle = hough_trans(circle_noise, hs_circle, coordinate='circle')
```
